```python
plt.figure(figsize=(8,8))

plt.subplot(1,4,1)
sns.boxplot(x=data['bedrooms'])
plt.title('Bedrooms')

plt.subplot(1,4,2)
sns.boxplot(x=data['bathrooms'])
plt.title('Bathrooms')

plt.subplot(1,4,3)
sns.boxplot(x=data['sqft_living'])
plt.title('SqFt Living')

plt.subplot(1,4,4)
sns.boxplot(x=data['price'])
plt.title('Price')

plt.tight_layout()
plt.show()
```

```python
consideredata = data[['sqft_living','bedrooms','bathrooms','price']]
consideredata
```

```python
def extract_outliers(column):
  Q1 = column.quantile(0.25)
  Q3 = column.quantile(0.75)
  IQR = Q3 - Q1
  lowerbound = Q1 - 1.5*IQR
  upperbound = Q3 + 1.5*IQR
  return (column < lowerbound) | (column > upperbound)

outlier_sqft = extract_outliers(consideredata['sqft_living'])
outlier_bedrooms = extract_outliers(consideredata['bedrooms'])
outlier_bathrooms = extract_outliers(consideredata['bathrooms'])
outlier_price = extract_outliers(consideredata['price'])
combinedoutliers = outlier_sqft | outlier_bedrooms | outlier_bathrooms | outlier_price
withoutliers = ~combinedoutliers
cleanedata = consideredata[withoutliers]
cleanedata
```

```python
cleanedata.describe()
```

```python
from sklearn.model_selection import train_test_split
X = cleanedata[['sqft_living','bedrooms','bathrooms']]
Y = cleanedata['price']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
```

```python
plt.scatter(Y_test,Y_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.plot([Y_test.min(),Y_test.max()],[Y_test.min(),Y_test.max()],color='red')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
Y_pred = model.predict(X_test)
mae = mean_absolute_error(Y_test,Y_pred)
mse = mean_squared_error(Y_test,Y_pred)
r2 = model.score(X_test,Y_test)
print(f'Mean Absolute Error: {mae},Mean Squared Error: {mse},R2:{r2}')
```

```python
sqft = float(input('Enter Sq.ft of House:'))
nbedrooms = int(input('Enter Number of Bedrooms:'))
nbathrooms = int(input('Enter Number of Bathrooms:'))

def price_predict(model,sqft,nbedrooms,nbathrooms):
  inputdata = pd.DataFrame({'sqft_living':[sqft],'bedrooms':[nbedrooms],'bathrooms':[nbathrooms]})
  predictedprice = model.predict(inputdata)
  return predictedprice[0]

price =  price_predict(model,sqft,nbedrooms,nbathrooms)
print(f'The Predicted House Price is: Rs{price:,.2f}')
```
